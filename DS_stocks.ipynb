```python
import yfinance as yf
import pandas as pd

# Create a ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Retrieve historical data for the maximum period and save it in a DataFrame
tesla_data = tesla.history(period="max")

# Reset the index and display the first five rows
tesla_data.reset_index(inplace=True)
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Use a static URL provided for the assignment
tesla_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Get the HTML data from the website
tesla_html_data = requests.get(tesla_url).text

# Parse the HTML using BeautifulSoup with the built-in html.parser
tesla_soup = BeautifulSoup(tesla_html_data, "html.parser")

# Find all tables on the page
tables = tesla_soup.find_all("table")

# Look for the table containing "Tesla Quarterly Revenue"
revenue_table = None
for table in tables:
    if "Tesla Quarterly Revenue" in table.get_text():
        revenue_table = table
        break

if revenue_table is None:
    print("Revenue table not found!")
else:
    # Create a list to store each row as a dictionary
    rows_list = []
    tbody = revenue_table.find("tbody")
    if tbody:
        for row in tbody.find_all("tr"):
            cols = row.find_all("td")
            if cols:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace("$", "").replace(",", "")
                rows_list.append({"Date": date, "Revenue": revenue})
                
        # Create a DataFrame from the list
        tesla_revenue = pd.DataFrame(rows_list)
        
        # Remove rows where the Revenue is an empty string
        tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
        print(tesla_revenue.tail())
    else:
        print("No table body found in the revenue table.")
```
